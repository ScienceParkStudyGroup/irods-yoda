# Retrieving genomics datasets in the ENA repository

The idea is to produce a table listing datasets, their file location and their corresponding conditions so that meta-analysis of genomic datasets can be done. 

This tabular file would be the input for a Snakemake pipeline to estimate gene expression from these datasets. 

Eventually, this tool should be compatible with the iRODS API in order for public and private datasets to be combined.
